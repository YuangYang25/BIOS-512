Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
p = ggplot(penguins, aes(x = body_mass_g)) +
    geom_histogram(color = 'steelblue',fill = 'steelblue3', bins = 37) +
    facet_wrap(~year,ncol = 1,scale = 'free_y') +
    scale_x_continuous(labels = comma_format()) +
    scale_y_continuous(breaks = seq(0, 7.5, by = 2.5)) +
    theme_minimal() +
    labs(y = 'Number of Penguins',  x = 'Body mass in grams',title = 'Histograms of penguin body mass for measurements taken from 2007-2009')+
    theme(strip.text=element_text(hjust = 0.925,size = 14,color = 'grey50',face = 'bold'))
    
p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
p = ggplot(penguins,aes(x= body_mass_g, y = sex)) + 
    geom_point(na.rm = T, size = 0.8, alpha = 0.5, position = position_jitter(height = 0.25, width = 0)) +
    facet_wrap(~species,ncol = 1) +
    scale_x_continuous(labels = comma_format(),breaks = seq(3500,5500,by = 1000)) +
    theme_minimal() +
    labs(y = '',  x = 'Body mass in grams')+
    theme(strip.text=element_text(hjust = 0.05,size = 14,color = 'grey50',face = 'bold'))
    
p
```
